# Criando funções 

```python
def greeting(name=0):
  if (name == 0):
    name = input('What is your name? ')
  print(f'Hello, {name} ! Nice to meet you')
```

```python
name = 'lassy'
greeting(name)
```

```python
def fullName():
  firstName = input("Enter your first name: ")
  middleName = input("Enter your middle name: ")
  name = firstName + ' ' + middleName
  print(name)
```

# Laços e Loops


```python
ages
```

```python
# FOR fora da função

def canDrive(age):
  if age >= 18:
    print(f'{age} years old, ARE allowed to drive')
  else:
    print(f'{age} years old, ARE NOT allowed to drive')

for age in ages:
  canDrive(age)
```

```python
# FOR dentro da função

def canDrive(ages):
  for age in ages:
     if age >= 18:
      print(f'{age} years old, ARE allowed to drive')
     else:
      print(f'{age} years old, ARE NOT allowed to drive')

canDrive(ages)
```

# Booleano

```python
permissions = []
ages = [20, 14, 40]

def canDrive(ages, permissions):
  for age in ages:
    if age >= 18:
      permissions.append(True)
    else:
      permissions.append(False)

canDrive(ages, permissions)
```

```python
permissions
```

```python
for permission in permissions:
  if permission == True:
    print('Are allowed to drive')
  else:
    print('Are not allowed to drive')
```
